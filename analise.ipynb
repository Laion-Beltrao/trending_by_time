# 3 - Criar colunas

```python
# criar coluna de mês
df['month'] = df['trending_date'].dt.month
```

```python
# criar coluna de semana
df['month'] = df['trending_date'].dt.isocalendar().week
```

```python
# Criar coluna de dias da semana
df['week_days'] = df['trending_date'].dt.to_period('W')
```
